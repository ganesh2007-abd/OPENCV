```python
import cv2
```

```python
img = cv2.imread("C:\\Users\\HP\\OneDrive\\Pictures\\mnheist.jpg")
img = cv2.resize(img,(860,620))
```

```python
cv2.imshow("IMG",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("IMG",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
canny_img = cv2.Canny(gray,150,200)
```

```python
cv2.imshow("IMG",canny_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

FOR VIDEOS

```python
vid = cv2.VideoCapture(0)

while True:
    try:
        res,frame = vid.read()
        g_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        c_frame = cv2.Canny(g_frame,150,200)

        cv2.imshow("Video",c_frame)
    except:
        pass
    if cv2.waitKey(3) & 0xff == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
import numpy as np
kernel = np.ones((3,3),np.uint8)
```

```python
erode_img = cv2.erode(canny_img,kernel,iterations=1)
```

```python
cv2.imshow("E_IMG",erode_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
dil_img = cv2.dilate(canny_img,kernel,iterations=1)
```

```python
cv2.imshow("D_IMG",dil_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
